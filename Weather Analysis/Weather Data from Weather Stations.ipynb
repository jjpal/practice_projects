# Data Dictionary Types
---
Non-numeric - categorical
---
* State : object 
* WS Airport Location  : object 
* Weather Station  : object 

Numeric - continuous
---
- Year   : int64
- T (°C)  - Average temperature in Celcius : float64
- T Avg (°F) - Average temperature in Farenheit : float64
- TM (°C) - Max Avg temperature in Celcius : float64
- T Avg Max (°F) - Max Avg temperature in Farenheit : float64
- Tm (°C) - Min Avg temperature in Celcius : float64
- T Avg Min (°F) - Min Avg temperature in Farenheit : float64
- PP - rain/snow precipation total : float64  
- V  - Avg wind speed : float64
- RA - Num days with rain : int64
- SN - Num days with snow : int64
- TS - Num of days with storms : int64
- FG - Num foggy days : int64
- TN -Number of days with tornado : int64
- GR - Num days with hail  : int64


```python
# Summary statistics of the nemeric features
df.describe()
```

```python
# Checking null values present in any of the rows
df.isnull().values.any()
```

```python
#To get information about data type of attributes
df.info()
```

```python
# drop rows with Average Temperature 0.0 - contains incomplete data to average
df.drop(df[df['T Avg (°F)'] <= 0.00].index, inplace = True)
df.shape                     # previously (1008, 18)
```

```python
# drop years are not shown for all states
df.drop(df[df['Year'] < 1975].index, inplace = True)
df.shape    
```

```python
df.sample(10)
```

```python
# plot numeric features
df.hist(figsize=(14,14), xrot=45)
plt.show()
```

```python
# Summary statistics of the categorical features
df.describe(include='object')
```

```python
# overwriting data after changing format
df['Year'] = pd.to_datetime(df["Year"], format="%Y")
df.info()
```

```python
# Group numeric features by each categorical feature
for column in df.select_dtypes(include='object'):
    if df[column].nunique() < 20:
        display(df.groupby(column).mean())
```

```python
#  Correlations matrix for the different numerical features
corrs = df.corr()
corrs
```

```python
# Heatmap of the correlations
plt.figure(figsize=(10,8))
sns.heatmap(corrs, cmap='RdBu_r', annot=True)
plt.show()
```

What to look out for:
---
Correlated features; 
> dark red (positive)
  - All the temperature related features as expected are strongly correlated
  - 'T (°C)', 'T Avg (°F)', 'TM (°C)', ' T Avg Max (°F)', 'Tm (°C)', 'T Avg Min (°F)'

> dark blue (negative)
 - Snow feature (SN) has a strongly negative relationship with the temperature features
